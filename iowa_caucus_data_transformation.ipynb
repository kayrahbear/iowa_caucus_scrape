```python
import pandas as pd
```

```python
csv = "iowa_caucus_results.csv"

scraped_data_df = pd.read_csv(csv)
```

```python
scraped_data_df.head()
```

```python
bennet_df = scraped_data_df.iloc[:,0:5]
bennet_df["Candidate"] = ['Michael Bennet' for i in range(1765)]
bennet_df = bennet_df.rename(columns={"Bennet_First_Expression": "first_expression", "Bennet_Final_Expression": "final_expression", "Bennet_SDE": "state_delegate_equivalent"})

bennet_df.head()
```

```python
biden_df = scraped_data_df.iloc[:,[0,1,5,6,7]]
biden_df["Candidate"] = ['Joe Biden' for i in range(1765)]
biden_df = biden_df.rename(columns={"Biden_First_Expression": "first_expression", "Biden_Final_Expression": "final_expression", "Biden_SDE": "state_delegate_equivalent"})

biden_df.head()
```

```python
bloomberg_df = scraped_data_df.iloc[:,[0,1,8,9,10]]
bloomberg_df["Candidate"] = ['Michael Bloomberg' for i in range(1765)]
bloomberg_df = bloomberg_df.rename(columns={"Bloomberg_First_Expression": "first_expression", "Bloomberg_Final_Expression": "final_expression", "Bloomberg_SDE": "state_delegate_equivalent"})

bloomberg_df.head()
```
